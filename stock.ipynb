```python
from crewai_tools import ScrapeWebsiteTool, SerperDevTool

# Initialize the tools
search_tool = SerperDevTool()
scrape_tool = ScrapeWebsiteTool()
```

```python
data_analyst_agent = Agent(
    role="Data Analyst",
    goal="Monitor and analyze {company} in real-time "
         "to identify trends and predict share movements.",
         llm = default_llm,
    backstory="Specializing in financial markets, this agent "
              "uses statistical modeling and machine learning "
              "to provide crucial insights on {company}. With a knack for data, "
              "the Data Analyst Agent is the cornerstone for "
              "informing trading decisions.",
    verbose=True,
    allow_delegation=True,
    tools = [scrape_tool, search_tool]
)
```

```python
# Task for Data Analyst Agent: Analyze Market Data
data_analysis_task = Task(
    description=(
        "Continuously monitor and analyze market data for "
        "the selected stock ({company}). "
        "Use statistical modeling and machine learning to "
        "identify trends and predict market movements."
    ),
    expected_output=(
        "Recommmendation on whether to buy/sell the {company} stock. And the recommended price to buy as well as sell"
          "to maximize the profits as per the market forecast and expert recommendations."
          "Also mention the source that is used to gather expert recommendations along with the number of experts naking the recommendation."
    ),
    agent=data_analyst_agent,
)
```

```python
# Define the crew with agents and tasks
financial_trading_crew = Crew(
    agents=[data_analyst_agent],
            tasks=[data_analysis_task]
)
```

```python
# Example data for kicking off the process
financial_trading_inputs = {
    'company': 'Muthoot Microfin Ltd'
}
```
